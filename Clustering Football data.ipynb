```python

from __future__ import division
import csv 
import pandas as pd
import numpy as np
import scipy.stats as scipy
import matplotlib.pyplot as plt
import requests #download from football data
import io 
import seaborn as sns
%matplotlib inline

from sklearn.cluster import KMeans
# preprocessing - normalises the data points
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
import bokeh.plotting as bp
```

```python
df = pd.read_csv("./Data/E0_17_18.csv")
```

```python
# iloc gets rows (or columns) at particular positions in the index (so it only takes integers).
result_16 = df.iloc[:,:23]
# Axis 0 will act on all the ROWS in each COLUMN
# Axis 1 will act on all the COLUMNS in each ROW
result_16 = result_16.drop(['Div','Date'],axis=1)

# 7 is FTR - so selects as far over as FTR
table_features = df.iloc[:,:7]
table_features = table_features.drop(['FTHG','FTAG','Div','Date'],axis=1)
# All beating columns
bet_16 = df.iloc[:,23:]
```

```python
# 0) Team, 1) Home Goals Score, 2) Away Goals Score, 3) Home Attack Strength 
# 4) Away Attack Strength 5) Home Goals Conceded, 6) Away Goals Conceded, 
# 7) Home Defensive Strenght, 8) Away Defensive Strength
# New table of columns created
table_16 = pd.DataFrame(columns=('Team','HGS','AGS','HAS','AAS','HGC','AGC','HDS','ADS',))
print(table_16)
```

```python
result_home = result_16.groupby('HomeTeam', as_index=True)
result_away = result_16.groupby('AwayTeam', as_index=True)
```

```python
result_home = result_16.groupby('HomeTeam', as_index=True).agg(np.mean)
result_home
```

```python
# scaling - normalises along the row of the dataset
# normalization - 
data = scale(result_home.values)
```

```python
# 4 is the number of 'k' clusters within the k-means algorithm
k = 4 
```

```python
kmeans = KMeans(n_clusters=k)
kmeans.fit(data)
clusters = pd.DataFrame(kmeans.cluster_centers_,columns=['FTHG','FTAG','HTHG','HTAG','HS','AS','HST','AST','HF','AF','HC','AC','HY','AY','HR','AR'])
result_home['TeamCluster'] = kmeans.labels_
print(result_home)
result_home.to_csv("testcluster.csv",sep=',')
```

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html
# PCA - Principal component analysis
# Linear dimensionality reduction using Singular Value Decomposition of the data to project it to a lower dimensional space.
# It uses the LAPACK implementation of the full SVD or a randomized truncated SVD by the method of Halko et al. 2009, depending on the shape of the input data and the number of components to extract.
clusters
tcluster = PCA(n_components=2).fit(clusters.values)
tcluster_2d = tcluster.transform(clusters.values)
tcluster_2d
```

```python
teamName = result_home.index 
```

```python
# Allows plotting to occur in 2 dimensions - components = 2, allows this to occur
tca = PCA(n_components=2).fit(result_home.values)
tca_2d = tca.transform(result_home.values)
```

```python
b = np.column_stack((teamName,tca_2d)).copy()
b
```

```python
for i in range(0, tca_2d.shape[0]):
    if kmeans.labels_[i] == 1:
        t1 = plt.scatter(tca_2d[i,0],tca_2d[i,1],c='r',marker='+')
    elif kmeans.labels_[i] == 0:
        t2 = plt.scatter(tca_2d[i,0],tca_2d[i,1],c='g',marker='o')
    elif kmeans.labels_[i] == 2:
        t3 = plt.scatter(tca_2d[i,0],tca_2d[i,1],c='b',marker='*')
    elif kmeans.labels_[i] == 3:
        t4 = plt.scatter(tca_2d[i,0],tca_2d[i,1],c='c',marker='s')
        
plt.title('K-means clustering into 4 clusters')
plt.show()
```
